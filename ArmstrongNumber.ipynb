```python
n= int(input("Enter a number:"))
nums_digit = 0
num =n
l=[]
sum=0
c=0
while(num>0):
  digit = num%10
  l.append(digit)
  c+=1
  num = num//10
for i in range(len(l)):
  sum+=(l[i])**c

if(sum==n):
  print("Number is Armstrong Number")
else:
  print("Number is not Armstrong Number")
```
