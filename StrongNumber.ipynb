```python
num = int(input("Enter Number"))
n = num
s=0
while(n>0):
  digit = n%10
  f=1
  for i in range(1,digit+1):
    f*=i
  s+=f
  n//=10
if(s==num):
  print("Strong Number")
else:
  print("Not a strong Number")
```
